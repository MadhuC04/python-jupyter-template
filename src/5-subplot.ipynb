### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create a 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Data for the plots
x_line = np.arange(0, 10, 1)
y_line = x_line ** 2

x_bar = ['A', 'B', 'C', 'D']
y_bar = [5, 7, 3, 9]

x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]

labels_pie = ['Python', 'Java', 'C++', 'JavaScript']
sizes_pie = [40, 25, 20, 15]

# Line chart in the first subplot
axs[0, 0].plot(x_line, y_line, marker='o', color='b')
axs[0, 0].set_title('Line Chart')
axs[0, 0].set_xlabel('x values')
axs[0, 0].set_ylabel('y = x^2')

# Bar chart in the second subplot
axs[0, 1].bar(x_bar, y_bar, color=['red', 'blue', 'green', 'purple'])
axs[0, 1].set_title('Bar Chart')
axs[0, 1].set_xlabel('Categories')
axs[0, 1].set_ylabel('Values')

# Scatter plot in the third subplot
axs[1, 0].scatter(x_scatter, y_scatter, color='orange')
axs[1, 0].set_title('Scatter Plot')
axs[1, 0].set_xlabel('x values')
axs[1, 0].set_ylabel('y values')

# Pie chart in the fourth subplot
axs[1, 1].pie(sizes_pie, labels=labels_pie, autopct='%1.1f%%', colors=['gold', 'lightcoral', 'lightskyblue', 'lightgreen'])
axs[1, 1].set_title('Pie Chart')

# Adjust layout for better spacing
plt.tight_layout()

# Display the plot
plt.show()
```
